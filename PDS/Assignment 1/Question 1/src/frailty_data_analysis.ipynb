Step 3: Data Analysis

```python
import os
import pandas as pd

df=pd.read_csv('/content/cleaned_frailty_data.csv')
df
```

```python
import seaborn as sbn
import matplotlib.pyplot as plt
#creates correlation matrix
correlation_matrix = df.corr()
# Create a heatmap
plt.figure(figsize=(8, 6))
sbn.heatmap(correlation_matrix, annot=True, cmap='cividis', linewidths=1.5)
plt.title("Heatmap")
plt.show()
```

```python
# histogram plot for Age vs Height
sbn.histplot(df, x='Age',y='Height',hue='Frailty')
plt.title("Age vs Height")
plt.xlabel="Age"
plt.ylabel="Height"
plt.show()
```

```python
# histogram plot for Age vs Weight
sbn.histplot(df, x='Age',y='Weight',hue='Frailty')
plt.title("Age vs Weight")
plt.xlabel="Age"
plt.ylabel="Weight"
plt.show()
```

```python
#piarplot
sbn.pairplot(df, hue='Frailty')
plt.show()
```

```python
#box plot
sbn.boxplot(x='Frailty', y='Grip strength', data=df)
plt.title("Grip strength vs Frailty")
plt.xlabel="Frailty"
plt.ylabel="Grip strength"
plt.show()
```
