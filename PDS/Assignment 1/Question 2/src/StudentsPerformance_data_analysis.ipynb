```
import os
import pandas as pd

df=pd.read_csv('/content/cleaned_StudentPerformance_data.csv')
df
```

```
import seaborn as sbn
import matplotlib.pyplot as plt
#creating correlation matrix
correlation_matrix = df.corr()
# Create a heatmap
plt.figure(figsize=(8, 6))
sbn.heatmap(correlation_matrix, annot=True, cmap='cividis', linewidths=1.5)
plt.title("Heatmap")
plt.show()
```

```
# generating pairplots between all the three scores
sbn.pairplot(df[['math score','reading score','writing score']])
plt.show()
```

```
# plot to get a count for the gender
plt.figure(figsize=(2, 6))
sbn.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```
# plot to get a count of differnt ethnic groups
plt.figure(figsize=(4, 6))
sbn.countplot(x='race/ethnicity', data=df)
plt.title("Race/Ethnicity distribution")
plt.show()
```

```
#Creating a pie chart for race/ethnicity
race_dist= df['race/ethnicity'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(race_dist, labels=race_dist.index, autopct='%1.1f%%')
plt.title('Race Distribution')
plt.show()
```

```
#Creating a pie chart to analyze paernts education level
parent_education= df['parental level of education'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(parent_education, labels=parent_education.index, autopct='%1.1f%%')
plt.title('Parent Level of Education')
plt.show()
```

```
#boxplots for gender vs math score
sbn.boxplot(x='gender', y='math score', data=df, hue='test preparation course')
plt.title("Gender vs Math Score")
plt.show()
```

```
# Writing score against test preparation course
sbn.histplot(x='writing score',data=df, hue='test preparation course', multiple='stack')
plt.title('Writing Score against test preparation course')
plt.show()
```

```
# scatter plot to writing score against test preparation course
sbn.scatterplot(x='writing score',y='reading score',data=df)
plt.title('Writing Score against test preparation course')
plt.show()
```
